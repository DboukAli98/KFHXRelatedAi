```python
#cleaning functions

import re
def remove_html_tags(text):
    """Remove html tags from a string"""
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def clean_KD(text):
    clean_kd = re.compile(r'\b\d+\s*KD\b')
    text = re.sub(clean_kd, '', text)

    return text.strip()
```

Data Cleaning

```python
deals['Description'] = deals['Description'].apply(remove_html_tags)
```

```python
deals["Title"] = deals["Title"].apply(clean_KD)
```
