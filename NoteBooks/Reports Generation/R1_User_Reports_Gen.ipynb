```python
#General R F M 
# Recency (Active / Inactive)
# Frequency (Total Transactions)
# Monetary (Total Amount Spent)

user_profiles['R'] = pd.qcut(user_profiles['recency'], 5, labels=[5, 4, 3, 2, 1])
user_profiles['F'] = pd.qcut(user_profiles['total_transactions'], 5, labels=[1, 2, 3, 4, 5])
user_profiles['M'] = pd.qcut(user_profiles['total_amount_spent'], 5, labels=[1, 2, 3, 4, 5])

# RFM Score on a scale of 1 to 5
user_profiles['RFM_Score'] = user_profiles[['R', 'F', 'M']].mean(axis=1).round().astype(int)

# Segmentation based on RFM score
def segment(x):
    if x == 5:
        return 'Champions'
    elif x == 4:
        return 'Loyal Customers'
    elif x == 3:
        return 'Potential Loyalist'
    elif x == 2:
        return 'New Customers'
    else:
        return 'At Risk'

user_profiles['RFM_Segment'] = user_profiles['RFM_Score'].apply(segment)
```

```python
user_profiles.columns = user_profiles.columns.astype(str)

mcc_columns = [col for col in user_profiles.columns if col.startswith('total_amount_mcc_')]

grouped = user_profiles.groupby('RFM_Segment')[mcc_columns].sum()


top_mccs_by_segment = {}
for segment in grouped.index:
    top_mccs = grouped.loc[segment].sort_values(ascending=False).head(5)  # Get top 5 MCCs
    top_mccs_by_segment[segment] = top_mccs

for segment, top_mccs in top_mccs_by_segment.items():
    print(f"Top MCCs for {segment}:")
    print(top_mccs)
    print()
```

```python
def is_numeric_column(col_name):
    try:
        int(col_name)
        return True
    except ValueError:
        return False
    
mcc_columns = [col for col in user_profiles.columns if is_numeric_column(col)]

mcc_amounts_cols = [col for col in user_profiles.columns.astype(str) if col.startswith('total_amount_mcc_')]
def calculate_user_interest_score(user_id):
    user_data = user_profiles[user_profiles['FK_BusinessUserId'] == user_id]
    
    total_frequencies = user_data[mcc_columns].sum(axis=1).values[0]
    total_paid_amount = user_data[mcc_amounts_cols].sum(axis=1).values[0]
    
    mcc_scores = {}
    if total_frequencies > 0 and total_paid_amount > 0:
        for mcc in mcc_columns:
            frequency = user_data[mcc].values[0]
            amount_col = f'total_amount_mcc_{mcc}' 
            print("amounttt colll " + amount_col)
            if amount_col in user_data.columns:
                print("enteredddd")
                amount = user_data[amount_col].values[0]
                if frequency > 0 and amount > 0:
                    score = 5 * (frequency / total_frequencies) * (amount / total_paid_amount)
                    mcc_scores[mcc] = score
    
    return mcc_scores
```
